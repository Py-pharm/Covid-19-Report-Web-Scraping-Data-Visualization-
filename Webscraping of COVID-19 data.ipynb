```python
# Importing the necessary libraries
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://covid19.ncdc.gov.ng/"
```

```python
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text, 'html.parser')
```

```python
table = soup.find('table', {'id':'custom1'})
```

```python
# Scraping the table headers
headers = []

for header in table.find_all('th'):
    title = header.text
    headers.append(title)
```

```python
headers
```

```python
# Appending the headers to the empty dataframe
df = pd.DataFrame(columns = headers)
```

```python
df
```

```python
# Scraping the remaining rows and appending them to the dataframe
for row in table.find_all('tr')[1:]:
    data = row.find_all('td')
    row_data = [td.text.strip() for td in data]
    length = len(df)
    df.loc[length] = row_data
```

```python
df
```

```python
# Manually getting other key information from the webpage
data_2 = {"samples rested":5441162, "Confirmed Cases":261473, "Active Cases":3373, "Discharged Cases":254953, "Death":3147}

```

```python
# Converting the dictionary to a dataframe
df2 = pd.DataFrame(data_2, index = [0])
```

```python
df2
```

```python
# Saving the two dataframes in csv formats
df.to_csv("complete report.csv")
df2.to_csv("key report.csv")
```
